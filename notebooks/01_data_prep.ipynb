## Data and metrics overview

This notebook prepares and inspects the data used in the robustness experiments.

- **Goal**: make sure the test set is correctly formatted (image + `.txt` with ground truth), quickly visualise it, and sanity‑check the CER/WER implementations.
- **Real data**: for a convincing study, you should eventually use **several hundred real images** (e.g. 300–1000 cropped words / lines) from scanned documents or a public dataset.
- **Synthetic demo**: the synthetic examples generated below are only for quick debugging; the main experiments should be run on your real `data/test/` set.

You can rerun this notebook after you replace the synthetic data with a real dataset to visually inspect its diversity and quality.

```python
# Quick sanity checks for CER / WER

examples = [
    ("hello world", "hello world"),
    ("hello world", "h3llo wor1d"),
    ("TOTAL: 123.45 USD", "TOTAL 123.45 USD"),
    ("invoice", "invoce"),
]

for ref, hyp in examples:
    print(f"REF: {ref!r}\nHYP: {hyp!r}\n  CER={cer(ref, hyp):.3f}  WER={wer(ref, hyp):.3f}\n")
```

```python
# Generate a small synthetic dataset (demo only)

out_dir = os.path.abspath("../data/test")
os.makedirs(out_dir, exist_ok=True)
create_synthetic_dataset(out_dir=out_dir, num_samples=8, seed=123)
print("Synthetic samples written to", out_dir)
```

```python
# Visualize a couple of synthetic images

import glob
import matplotlib.pyplot as plt

img_paths = sorted(glob.glob(os.path.join(out_dir, "synthetic_*.png")))[:4]

fig, axes = plt.subplots(1, len(img_paths), figsize=(12, 3))
for ax, p in zip(axes, img_paths):
    img = load_image_as_rgb(p)
    ax.imshow(img)
    ax.set_title(os.path.basename(p))
    ax.axis("off")

plt.tight_layout()
plt.show()
```
